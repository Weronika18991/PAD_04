```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dan/orders.csv')
df = pd.DataFrame(data)
df['order_date'] = pd.to_datetime(df['order_date'])
df
```

```python
# 2
unique = df['tshirt_category'].unique()
print(unique)
print("Length: ", len(unique))
```

```python
# 3

def name_normalizer(name, replacer, stringToReplace):
  normalized = []
  for i in name.split(" "):
    if i == replacer:
      normalized.append(stringToReplace)
    else:
      normalized.append(i)

  return " ".join(normalized)

df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalizer(x, 'Wh', 'White'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalizer(x, 'Tshirt', 'T-Shirt'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalizer(x, 'Bl', 'Black'))
df.tshirt_category = df.tshirt_category.apply(lambda x: name_normalizer(x, 'Tennis', 'T'))
df.tshirt_category.unique()
```

```python
# 4

def recognize(name, match):
  for i in name.split():
    if i in match:
      return i
  return 'unknown'

tshirt_colours = ("Black", "White")
tshirt_types = ("T-Shirt", "Shirt", "Hoodie")
tshirt_sizes = ("M", "F")

df["tshirt_type"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_types))
df["tshirt_size"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_sizes))
df["tshirt_colour"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_colours))
df
```
